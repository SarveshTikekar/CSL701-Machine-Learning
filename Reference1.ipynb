```python
def linear_regression(a, b):
  # Perform linear regression
  slope, intercept, r_value, p_value, std_err = linregress(a, b)

  # Create regression line points
  regression_line = [slope * x + intercept for x in a]

  # Plotting
  plt.figure(figsize=(8, 6))
  plt.scatter(a, b, label='Data points')
  plt.plot(a, regression_line, color='red', label='Linear regression line')

  print("r_value is",r_value)
  print("p_value is",p_value)
  print("std_err is", std_err)
```

## Values returned by the linregress function:

**r_value:**
* Also known as the Pearson correlation coefficient, r_value measures the strength and direction of the linear relationship between a and b. It ranges from -1 to 1:
   * r_value near +1 indicates a strong positive linear relationship.
   * r_value near -1 indicates a strong negative linear relationship.
   * r_value near 0 indicates a weak or no linear relationship.

* r_value is useful for determining how well the linear regression model fits the data points.

**p_value:**
* The p-value is a measure of the probability that the observed correlation (r_value) could occur due to random chance, assuming there is actually no relationship between a and b in the population.
* A low p-value (typically less than 0.05) suggests that the observed correlation is statistically significant and not likely to be due to random chance.
* If the p-value is high (greater than 0.05), it indicates that the observed correlation could plausibly be due to random variation, and the relationship may not be statistically significant.

**std_err:**
* The std_err (standard error of the regression) is a statistical measure that indicates the average amount that actual values of the dependent variable (y) deviate from the predicted values (on the regression line).


# Part 2: mtcars dataset

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset
file_path = '/content/mtcars.csv'
mtcars = pd.read_csv(file_path)
```

Dataset column values:
* mpg - 	Miles/(US) gallon

* cyl - 	Number of cylinders

* disp - 	Displacement (cu.in.)

* hp - 	Gross horsepower

* drat - 	Rear axle ratio

* wt - 	Weight (1000 lbs)

* qsec - 	1/4 mile time

* vs - 	Engine (0 = V-shaped, 1 = straight)

* am - 	Transmission (0 = automatic, 1 = manual)

* gear - 	Number of forward gears

```python
# Display summary of the dataset
summary = mtcars.describe()
info = mtcars.info()

print("Summary Statistics:")
print(summary)

print("\nDataset Information:")
print(info)
```

```python
# Display the first few rows of the dataset to understand its structure
print(mtcars.head())
```

```python
# Define the dependent and independent variables
a = mtcars['hp']  # Independent variable
b = mtcars['mpg']  # Dependent variable

# Define the linear regression function
def linear_regression(a, b):
    # Perform linear regression
    slope, intercept, r_value, p_value, std_err = linregress(a, b)

    # Create regression line points
    regression_line = [slope * x + intercept for x in a]


    plt.figure(figsize=(8, 6))
    plt.scatter(a, b, label='Data points')
    plt.plot(a, regression_line, color='red', label='Linear regression line')
    plt.xlabel('Horsepower (hp)')
    plt.ylabel('Miles per Gallon (mpg)')
    plt.title('Linear Regression: mpg vs hp')
    plt.legend()
    plt.show()

    print("r_value is", r_value)
    print("p_value is", p_value)
    print("std_err is", std_err)

# Call the linear regression function with the mtcars data
linear_regression(a, b)
```
